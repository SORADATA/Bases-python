```python
import requests
import pandas
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
def build_request_nominatim(country, city):
    """
    Construit une URL pour interroger l'API Nominatim avec une requête structurée
    
    Args:
        country (str): Nom du pays
        city (str): Nom de la ville
        
    Returns:
        str: URL complète de la requête
    """
    base_url = "https://nominatim.openstreetmap.org/search"
    # Utiliser les paramètres 'country' et 'city' comme demandé dans la Structured Query
    params = f"country={country}&city={city}&format=json"
    url_request = f"{base_url}?{params}"
    
    return url_request


```

```python

url_request_nominatim = build_request_nominatim("France", "Montrouge")
url_request_nominatim
```

```python
def get_lat_long(query):
     # IMPORTANT : Nominatim exige un User-Agent
    headers = {
        'User-Agent': 'FormationPython/1.0 (sissokomoussa611@gmail.com)'
    }
    response = requests.get(query)
    code = response.status_code
    if code == 200:
        json_data = response.json()
        latitude = float(json_data[0]["lat"])
        longitude = float(json_data[0]["lon"])
        return latitude, longitude
    else:
        print(f"Error code : {code}")
        return None
```
