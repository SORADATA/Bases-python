## Partie 2 : Nettoyage des données

```python
# Val manquantes dans la col des prénoms

df_prenoms[df_prenoms["preusuel"].isnull()]
```

```python
# Proportion des id avec _prenoms rare

prop_xxx = df_prenoms.groupby('preusuel')["nombre"].sum()["_PRENOMS_RARES"] / df_prenoms["nombre"].sum()
prop_xxx
```

```python
df_prenoms = df_prenoms.replace("_PRENOMS_RARES", np.nan)
df_prenoms
```

```python
## Supprimer les nan

df_prenoms = df_prenoms.replace('XXXX', np.nan)


df_prenoms = df_prenoms.dropna()

df_prenoms
```

```python
df_prenoms = df_prenoms.replace('XXXX', np.nan)
df_prenoms
```

```python
df_prenoms['annais'] = pd.to_numeric(df_prenoms['annais'])
df_prenoms['sexe'] = df_prenoms['sexe'].astype('category')
```
