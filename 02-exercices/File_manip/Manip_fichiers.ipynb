```python
# Pour ouvrir un ficher 
path_gamme = Path.cwd() /'gamme.txt'
file_in = open(path_gamme,'r')
file_in
file_in.readlines() # Affiche moi les élements de contient gamme.txt
file_in.close() # Une fois les manip terminées on ferme la connexion via la méthode close
```

```python
# Bonnes pratiques afin de fermer la connexion 

with open(path_gamme, 'r') as file_in:
    lines =file_in.readlines()
lines
# Cette facon de faire assure que lorsqu'on se trouve à la ligne initiale il arrete directement la connexion 
```

```python
# Lire un fichier
with open (path_gamme, 'r') as file_in:
    for line in file_in:
        print(line)
```

```python
with open(path_gamme, 'r') as file_in:
    txt = file_in.read()
txt
```

```python
# ecrire dans un fichier

ex = ["ceci", "est", "un", "exemple"]

with open("test.txt", "w") as file_out:
    for elem in ex:
        file_out.write(elem)
```

```python
!ls
```

```python
# Vérification du contenu 

with open("test.txt", "w") as file_out:
    for elem in ex:
        file_out.write(elem + "\n")

with open("test.txt", "r") as file_out:
    print(file_out.read())
```
